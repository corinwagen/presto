```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.rc("font", family="serif", size=11)
plt.rc("axes", titlesize=11)
plt.rc("axes", labelsize=11)
```

```python
df = np.genfromtxt("energy_equil.txt",delimiter=",",skip_header=1)

NMOLS = 100

times = df[:,0]
TE = df[:,1] / NMOLS
KE = df[:,2] / NMOLS
PE = df[:,3] / NMOLS
temp = df[:,4]
```

```python
LW = 0.3
AL = 0.8

f = plt.figure(figsize=(16,8))

a1 = f.gca()
a1.plot(times, TE - 1*np.mean(TE), color="black",label="total energy", linewidth=LW, alpha=AL)
a1.plot(times, KE - 1*np.mean(KE), color="blue",label="kinetic energy", linewidth=LW, alpha=AL)
a1.plot(times, PE - 1*np.mean(PE), color="green",label="potential energy", linewidth=LW, alpha=AL)
a1.set_xlabel("Time (fs)")
a1.set_ylabel("Energy per Molecule (kcal/mol)")
a1.set_xlim(left=0, right=100000)
a1.set_ylim(top=15, bottom=-5)
a2 = a1.twinx()
a2.plot(times, temp, color="red", label="temperature", linewidth=LW, alpha=AL)
a2.set_ylabel("Temperature (K)")
a2.set_ylim(top=550, bottom=150)

f.legend(loc=(0.8,0.1))

plt.title("Energy Conservation Over 100 ps")
plt.show()
```

```python
df = np.genfromtxt("energy_long.txt",delimiter=",",skip_header=1)

NMOLS = 100

times = df[:,0]
TE = df[:,1] / NMOLS
KE = df[:,2] / NMOLS
PE = df[:,3] / NMOLS
temp = df[:,4]
```

```python
LW = 0.3
AL = 0.8

f = plt.figure(figsize=(16,8))

a1 = f.gca()
a1.plot(times, TE - 1*np.mean(TE), color="black",label="total energy", linewidth=LW, alpha=AL)
a1.plot(times, KE - 1*np.mean(KE), color="blue",label="kinetic energy", linewidth=LW, alpha=AL)
a1.plot(times, PE - 1*np.mean(PE), color="green",label="potential energy", linewidth=LW, alpha=AL)
a1.set_xlabel("Time (fs)")
a1.set_ylabel("Energy per Molecule (kcal/mol)")
a1.set_xlim(left=0, right=100000)
a1.set_ylim(top=15, bottom=-5)

a2 = a1.twinx()
a2.plot(times, temp, color="red", label="temperature", linewidth=LW, alpha=AL)
a2.set_ylabel("Temperature (K)")
a2.set_ylim(top=550, bottom=150)

f.legend(loc=(0.8,0.1))

plt.title("Energy Conservation Over 100 ps")
plt.show()
```

```python
# what is the linearized drift rate

cutoff = 10000

slope, intercept = np.polyfit(times[cutoff:], TE[cutoff:] * NMOLS, 1)

slope = slope * 1000 # fs/ps

print(f"ENERGY DRIFT: {slope:.5f} kcal/(molps)")
```
