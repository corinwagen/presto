```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

SMALL_SIZE = 11
MEDIUM_SIZE = 12
BIGGER_SIZE = 14

plt.rc('font', size=SMALL_SIZE, family="serif")          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
ps5_df = np.genfromtxt("pes_5ps.csv", skip_header=1)
ps5_ccl_dist = ps5_df[:,0]
ps5_energy = ps5_df[:,1]

ps10_df = np.genfromtxt("pes_10ps.csv", skip_header=1)
ps10_ccl_dist = ps10_df[:,0]
ps10_energy = ps10_df[:,1]
```

```python
f = plt.figure(figsize=(8,6))
plt.plot(ps5_ccl_dist, ps5_energy, color="blue")
plt.plot(ps10_ccl_dist, ps10_energy, color="green")

plt.title("Free Energy Surface for C–Cl Dissociation")
plt.xlabel("C1–Cl7 Distance (Å)")
plt.ylabel("∆G (kcal/mol)")
plt.show()
```
